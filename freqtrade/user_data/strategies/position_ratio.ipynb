```python
window = 120  # 可以用30天、60天等
df["m2_rolling_mean"] = df["global_m2"].rolling(window=window).mean()
df["m2_rolling_std"] = df["global_m2"].rolling(window=window).std()
df["m2_zscore"] = (df["global_m2"] - df["m2_rolling_mean"]) / df["m2_rolling_std"]
df["m2_zscore"].corr(df["price_change"].shift(-1), method="spearman")
```

```python
df['global_m2'].corr(df['Price'].shift(-1), method="spearman")
```

```python
# 首先正确定义目标变量
df["btc_return_next_day"] = df["Price"].pct_change().shift(-1)  # 次日收益率

# 方法1：最直接的构造方法（应该优先尝试）
df["predicted_return"] = (df["global_m2"] - df["Price"]) / df["Price"]
corr_direct = df["predicted_return"].corr(df["btc_return_next_day"])
print(f"直接差值法相关性: {corr_direct:.4f}")

# 方法2：使用原始值计算，然后标准化结果
df["predicted_return_raw"] = (df["global_m2"] - df["Price"]) / df["Price"]
# 对最终预测值进行标准化，而不是对输入标准化
df["predicted_return_normalized"] = (df["predicted_return_raw"] - df["predicted_return_raw"].rolling(60).mean()) / df[
    "predicted_return_raw"
].rolling(60).std()

corr_normalized = df["predicted_return_normalized"].corr(df["btc_return_next_day"])
print(f"结果标准化法相关性: {corr_normalized:.4f}")

# 方法3：比率法（更稳健）
df["m2_price_ratio"] = df["global_m2"] / df["Price"]
df["predicted_return_ratio"] = df["m2_price_ratio"] - df["m2_price_ratio"].shift(1)  # 比率的变化

corr_ratio = df["predicted_return_ratio"].corr(df["btc_return_next_day"])
print(f"比率变化法相关性: {corr_ratio:.4f}")
```
